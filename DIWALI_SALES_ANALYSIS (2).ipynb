# Exploratory Data Analysis

#### Gender

```python
df.columns
```

```python
#Creating a count plot of "Gender" using Seaborn
ax = sns.countplot(data=df,x="Gender",hue="Gender")

#Add Labels to the Bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
#Grouping the DataFrame by "Gender" calculating the total "Amount" for the each gender, sorting the result by "Amount" in Descending Order

sales_by_Gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_by_Gen
```

```python
# Creating Pai Chart to visualize total sales by Gender
plt.pie(sales_by_Gen["Amount"],labels=sales_by_Gen["Gender"],autopct="%1.1f%%")
plt.title("Total Sales By Gender")
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.show()
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

#### Occupation

```python
# Set the figure size for the plot
sns.set(rc={"figure.figsize":(18,5)})

#Creating a count plot to visualize to distribution of "Occupation"
ax = sns.countplot(data=df,x="Occupation",hue="Occupation")

#Add labels to the bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Distribution of Occupation")
plt.xlabel("Occupation")
plt.show()
```

```python
# Grouping the DataFrame by "Occupation" calculating the total "Amount" for each Occcupation amd sorting the values
sales_status = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.set(rc={"figure.figsize":(15,5)}) # Setting the figure size for the plot

# Creating a bar plot to visualize the total 'Amount' for each occupation
ax = sns.barplot(sales_status, x="Occupation",y="Amount",hue="Occupation")
    
plt.title("Sales_Status")
plt.xlabel("Occupation")
plt.ylabel("Amount")
plt.show()
```

#### Age

```python
# Creating a count plot to visualize Age Group distribution with Gender compariosn
ax = sns.countplot(data=df,x="Age Group",hue="Gender")
# Add labels to the bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Total Amount vs Age Group
sales_by_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set_theme()
sns.barplot(x="Age Group",y="Amount",hue="Age Group", data=sales_by_age)
plt.show()
```

#### Marital Status

```python
#Create a count of value for each category of Marital Status
marital_count=df["Marital_Status"].value_counts()

#Plot a pai chart using the marital ststus
plt.figure(figsize=(10,5))
plt.pie(marital_count,labels=marital_count.index,autopct="%1.1f%%")
plt.title("Distribution of Marital Status")
plt.show()
```

```python
# Group the DataFrame by 'Marital_Status' and 'Gender', calculate the total 'Amount' for each combination, and sort the values
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)}) # Set the figure size for the plot

# Creating a bar plot to visualize 'Amount' based on 'Marital_Status' with comparison by 'Gender'
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender') 
```

```python
# Group the DataFrame by "Marital_Status" and "Gender" , calculate the total "Amount" for each combination , and sort the values
sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

#set the figure size for the plot
sns.set(rc={"figure.figsize":(5,5)})

#creating a bar plot to visualize "Amount" based on "Marital_Status" with comparison by "Gender"
sns.barplot(data=sales_state,x="Marital_Status",y="Amount",hue="Gender")

plt.show()
```

#### Product Category

```python
sns.set(rc={"figure.figsize":(27,7)})

ax= sns.countplot(data=df,x="Product_Category",hue="Product_Category")

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# Grouping the DataFrame by 'Product_Category', calculating the total 'Amount' for each category, and sorting the values
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)}) # Setting the figure size for the plot
# Creating a bar plot to visualize the total 'Amount' for each 'Product_Category'
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```

```python
# Grouping the DataFrame by "Product Category" calculating the total "Amount", and sorting the values
sales_by_category=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)

sns.set(rc={"figure.figsize":(20,5)})

sns.barplot(data=sales_by_category,x="Product_Category",y="Amount",hue="Product_Category")

plt.show()
```

### Married women in the age group of 26-35 years from Uttar Pradesh, Maharashtra, and Karnataka, working in IT, Healthcare, and Aviation, are more inclined to purchase products from the Food, Clothing, and Electronics categories.

#### State

```python
# Group the DataFrame by "State", calculate the sum of "Orders" ,and sort to get top 10 states with higest Orders
Orders_by_State=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)

# Set the figure size for th plot
sns.set(rc={"figure.figsize":(20,5)}) 

#Creating bar plot to visualize the top 10 Sates with th higest orders
sns.barplot(data=Orders_by_State,x="State",y="Orders",hue="State")

plt.show()
```

```python
# Group the DataFrame by 'State', calculate the sum of 'Amount', and extract thr top 10 states with highest total Amount
sales_state = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)

sns.set(rc={"figure.figsize":(15,5)}) # Set the figure size for the plot

sns.barplot(data=sales_state, x="State", y="Amount",hue="State") # Creating bar plot to visualize the top 10 states with highest orders

plt.show()
```
